The * in a function call "unpacks" a list (or other iterable), making each of its elements a separate argument. So without the *, you're doing zip( [[1,2,3],[4,5,6]] ). With the *, you're doing zip([1,2,3], [4,5,6])

```python
def one_hot_encode(X, K):
    # Input: (N, D)
    # Output: turn into one_hot rating (0.5, 1, 1.5 ...) --> (1 ,2, 3..10) --> (0, 1, ... 9)
    N, D = X.shape
    Y = np.zeros((N, D, K))
    for n, d in zip(*X.nonzero()):
        k = int(X[n, d] * 2 - 1)  # (0, 1, 2,...9)
        Y[n, d, k] = 1
    return Y
```

```python
def one_hot_mask(X, K):
    # doing same thing as one-hot-encode
    # if rating X is 0.5 then --> [1,0,0,0,0..0]
    # our mask is all ones [1,1,1,1..1]
    # When we X * Mask, still == X
    # if X is 0, Mask is all zeros [0,0,0...0] --> then lost * Mask == [0,0,0..0]
    N, D = X.shape
    M = np.zeros((N, D, K))
    for n, d in zip(*X.nonzero()):
        M[n, d, :] = 1
    return M
```

#### predicted rating = 0.5 * sum( probs * one_to_ten )  --> (0.5 ~ 5)
probs = [0.1, 0.2, 0.7]  ;  one_to_ten = [1, 2, 4]

```python
one_to_ten = np.arange(10) + 1  # array 1 ~ 10
def convert_porbs_to_ratings(probs):
    return probs.dot(one_to_ten / 2)
```

#### Two useful Tensor Multiplications : 
- since we working with tensors of dim>2, can't use np.matmul() <br>
- And we need **tf.tensordot(axes=[...])** to specify which dim Multiply with which dim

```python
def dot1(V, W):
    # Visible data --> Hidden
    # Input: V is (N, D, K)
    # Input: W is (D, K, M) .....note that dot1(V, b) works as well, since b is (D, K) 
    #  N is user batch_size; D is num_movie; 
    #  K is the one-hot dim (0~9 ratings); M is the Hidden layer dim
    # Output: h is (N, M)
    return tf.tensordot(V, W, axes=[[1, 2], [0, 1]])
```

```python
def dot2(H, W):
    # Hidden --> Visible data
    # H is (N, M) batches of hiddens
    # W is (D, K, M)
    # Output is the V (N, D, K)  batches of visible
    return tf.tensordot(H, W, axes=[[1], [2]])
```

## Start building RMB ! 

```python
class RBM(object):
    def __init__(self, D, M, K):
        self.D = D  # input dim (here nun_movie)
        self.M = M  # hidden layer dim 
        self.K = K  # possible number of rating (0.5 ~ 5) (K = 10)
        self.build(D, M, K)
        
    def build(self, D, M, K):
        # initialize params
        self.W = tf.Variable(initial_value=tf.random_normal(shape=(D, K, M)) * np.sqrt(2.0 / M))
        self.b = tf.Variable(initial_value=np.zeros((D, K)).astype(np.float32))
        self.c = tf.Variable(initial_value=np.zeros((M)).astype(np.float32))
        
        # data
        self.X_in = tf.placeholder(tf.float32, shape=(None, D, K))
        self.mask = tf.placeholder(tf.float32, shape=(None, D, K))
        
        # conditional probabilities
        # From V -> Hidden
        V = self.X_in
        p_h_given_v = tf.nn.sigmoid(dot1(V, self.W) + self.c)  # using Dot1() func
        self.p_h_given_v = p_h_given_v  # save for later use
        
        # Draw samples from p(h | v)
        r = tf.random_uniform(shape=tf.shape(p_h_given_v))
        H = tf.to_float(r < p_h_given_v)
        
        # Sample V' given sampled H
        # just sample from p(V' | H)
        logits = dot2(H, self.W) + self.b  # logits, since V' has 10-dims for one-hot Ratings
        cdist = tf.distributions.Categorical(logits=logits)  # using the predicted Prob
        X_sample = cdist.sample()  # sample --> shape is (N, D)
        X_sample = tf.one_hot(X_sample, depth=self.K)  # turn it into (N, D, K)
        X_sample = X_sample * self.mask  # the whole training will have the same mask 
        
        # The Objective :  minimize the diff bte free_energy of true values X_in and free_energy of X_sample
        objective = tf.reduce_mean(self.free_energy(self.X_in) - tf.reduce_mean(self.free_energy(X_sample)))
        self.train_op = tf.train.AdamOptimizer(learning_rate=0.005).minimize(objective)
                      
        # build the cost
        # we won't use this to optimize the model parameters
        # just to observe what happens during training, will the Cross-Entropy go down ? 
        logits = self.forward_logits(self.X_in)
        self.cost = tf.reduce_mean(
            tf.nn.softmax_cross_entropy_with_logits(
                labels=self.X_in, 
                logits=logits
            )
        )
        
        # to get the output
        self.output_visible = self.forward_output(self.X_in)
        init_op = tf.global_variables_initializer()
        self.sess = tf.Session()
        self.sess.run(init_op)
    
    
    # Fitting: The Training Part!
    def fit(self, X, mask, X_test, mask_test, epochs=10, batch_sz=256, show_fig=True):
        
        N, D = X.shape
        num_batch = N // batch_sz
        
        costs = []
        test_costs = []
        for e in range(epochs):
            t0 = time()
            print('Epoch: ', e)
            # Remember to shuffle test, training data all in the same order --> b/c we need to compute cost
            # sklearn.utils.shuffle(A, B) will shuffle A,B (having same dim_0) in the same order! 
            X, mask, X_test, mask_test = shuffle(X, mask, X_test, mask_test)
            
            for b in range(num_batch):
                split_idx = b * batch_sz
                #  .toarray() turn into a Dense function
                # Not necessary, but the training can be so much faster
                x = X[split_idx: split_idx+batch_sz].toarray()
                m = mask[split_idx: split_idx+batch_sz].toarray()
                
                # both visible units and mask have to be in one-hot form
                # N x D --> N x D x K
                x_batch_one_hot = one_hot_encode(x, self.K)
                m = one_hot_mask(m, self.K)
                
                # record cost
                _, c = self.sess.run(
                    (self.train_op, self.cost), 
                    feed_dict={self.X_in: x_batch_one_hot, 
                               self.mask: m}
                )
                
                if b % 10 == 0:
                    print("Curr_b / num_batches:", b, "/", num_batch, "cost:", c)
            
            print('duration: %.1f minutes' %((time()-t0)/60.0))
            
            # calculate the true train and test cost
            
            t0 = time()
            sse = 0
            test_sse = 0
            n = 0
            test_n = 0
            for b in range(num_batch):
                split_idx = b * batch_sz
                x = X[split_idx: split_idx+batch_sz].toarray()
                m = mask[split_idx: split_idx+batch_sz].toarray()
                # only visible input has to be in one-hot form
                x_oh_train = one_hot_encode(x, self.K)
                
                probs = self.get_visible(x_oh_train)
                pred_x = convert_porbs_to_ratings(probs)  # predicted rating: 0.5 ~ 5
                
                # Training Error
                sse += (m * (pred_x - x) * (pred_x - x)).sum()
                n += m.sum()
                
                # Test Error
                # the test PREDICTIONS come from the train data!
                # X_test and mask_test are only used for targets
                xt = X_test[split_idx: split_idx+batch_sz].toarray()
                mt = mask_test[split_idx: split_idx+batch_sz].toarray()
                test_sse += (mt * (pred_x - xt) * (pred_x - xt)).sum()
                test_n += mt.sum()
                
            c = sse/n
            ct = test_sse/test_n
            print("train mse:", c)
            print("test mse:", ct)
            print('calculate cost duration: %.1f minutes' %((time()-t0)/60.0))
            costs.append(c)
            test_costs.append(ct)
        
        # After finish ALL Epcohs, print!
        if show_fig:
            plt.plot(costs, label='train mse')
            plt.plot(test_costs, label='test mse')
            plt.legend()
            plt.show()
            
    def free_energy(self, V):
        first_term = - tf.reduce_sum(dot1(V, self.b))
        second_term = - tf.reduce_sum(
            tf.nn.softplus(dot1(V, self.W) + self.c), 
            axis=1
            )
        # tf.nn.softplus = log(exp(...) + 1) !!  
        # since H can take only binary values 0/1 
        #       --> exp(1*features) + exp(0*features) == exp(features) + 1
        return first_term + second_term
    
    def forward_hidden(self, X):
        # Input -> Hidden
        return tf.nn.sigmoid(dot1(X, self.W) + self.c)
    def forward_logits(self, X):
        # Input -> Hidden -> V's logits
        Z = self.forward_hidden(X)
        return dot2(Z, self.W) + self.b
    def forward_output(self, X):
        # Input -> Hidden -> V' logits -> Softmax Rating
        return tf.nn.softmax(logits=self.forward_logits(X))
    
    def transform(self, X):
        # accepts and returns a real numpy array
        # unlike forward_hidden and forward_output
        # which deal with tensorflow variables
        return self.sess.run(self.p_h_given_v, feed_dict={self.X_in: X})
                                   
    def get_visible(self, X):
        return self.sess.run(self.output_visible, feed_dict={self.X_in: X})
    
        
```

```python
def get_it_RUN():
    A = load_npz("./data/Atrain.npz")
    A_test = load_npz("./data/Atest.npz")
    mask = (A > 0) * 1.0
    mask_test = (A_test > 0) * 1.0

    N, M = A.shape
    rbm = RBM(M, 50, 10)
    rbm.fit(A, mask, A_test, mask_test)
```

```python
get_it_RUN()
```
